
Printar tela board

Escolher X ou O

Começa jogo, alternando turnos das escolhas de jogador 1 e 2:

A cada turno, checa-se se a um vencedor ou empate

Se houver vencedor, Display Vencedor

Se nao, Display Empate

Gostaria de jogar novamente?

```python
def display_board(p_list):

    clear_output()
    
    print( "          |       |   ")
    print( f"      {p_list[6]}   |   {p_list[7]}   |   {p_list[8]}")
    print( "          |       |   ")
    print( 29*"=")
    print( "          |       |   ")
    print( f"      {p_list[3]}   |   {p_list[4]}   |   {p_list[5]}")
    print( "          |       |   ")
    print( 29*"=")
    print( "          |       |   ")
    print( f"      {p_list[0]}   |   {p_list[1]}   |   {p_list[2]}")
    print( "          |       |   ")
```

```python
def player_choice():
    player = ""
    while player not in ["X","O"]:
        player = input("Gostaria de jogar com X ou O?")
        player = player.upper()

        if player not in ["X","O"]:
            print("Opção inválida!")
    return player
```

```python
def player_position(p_list,turn):
    user_choice = "wrong"
    
    while not user_choice.isdigit():
        if turn%2 != 0:
            user_choice = input("Jogador 1, escolha uma posição a inserir X: ")
        else:
            user_choice = input("Jogador 2, escolha uma posição a inserir O: ")
        if not user_choice.isdigit():
            print("Posição inválida!")
            continue
            
        if int(user_choice) not in p_list:
            print("Posição inválida!")
            user_choice = "Wrong!"        
    return int(user_choice)
```

```python
def win_check(p_list, mark):
  
    for i in mark:
        if p_list[0] == p_list[1] == p_list[2] == i:
            return i   
        elif p_list[3] == p_list[4] == p_list[5] == i:
            return i  
        elif p_list[6] == p_list[7] == p_list[8] == i:
            return i  
        elif p_list[0] == p_list[4] == p_list[8] == i:
            return i  
        elif p_list[2] == p_list[4] == p_list[6] == i:
            return i 
        elif p_list[0] == p_list[3] == p_list[6] == i:
            return i  
        elif p_list[1] == p_list[4] == p_list[7] == i:
            return i  
        elif p_list[2] == p_list[5] == p_list[8] == i:
            return i  
```

```python
def draw_check(board):
    for item in board:
        if type(item) is int:
            draw = False
            break
        else:
            draw = True 
    return draw
```

```python
def gameon_choice():
    choice = "Wrong!"
    choice_list = ["Y","N"]

    while choice not in choice_list:
        choice = input("Gostaria de continuar jogando? Y ou N")
        choice = choice.upper()
        if choice not in choice_list:
            print("Erro, escolha Y ou N")

    if choice == "Y":
        return True
    else:
        return False 
```

```python
# TIK TAK TOE MILESTONE 1 (MINHA RESOLUÇÃO ORIGINAL, SEM CONSULTA A SOLUÇÃO)
from IPython.display import clear_output
game_on = True

while game_on:
    # DEFINE O BOARD INICIAL
    board = list(range(1,10))
    # MOSTRA O BOARD INICIAL
    display_board(board)
    # JOGADOR ESCOLHE X OU O (NA MINHA CULTURA O "X" SEMPRE COMEÇA, TORNANDO ESSA FEATURE INÚTIL)
    player = player_choice()
    # INICIA CONTAGEM DE TURNOS
    turn = 1

    while game_on:
        # JOGO COMEÇA COM O POSICIONAMENTO (PRIMEIRO X)
        new_position = player_position(board,turn)
        # CHECA O TURNO PARA ALOCAR X OU O
        if turn%2 != 0:
            board[new_position-1] = "X"
        else:
            board[new_position-1] = "O"
        # MOSTRA O BOARD COM ALTERAÇAO
        display_board(board)
        # CHECA SE HÁ VENCEDOR
        winner = win_check(board,["X","O"])
        # CHECA SE HÁ EMPATE
        draw = draw_check(board)
        if winner == "X":
            print("Parabéns jogador 1, você ganhou!")
            game_on = False
        elif winner == "O":
            print("Parabéns jogador 2, você ganhou!")
            game_on = False
        elif draw:
            print("Empate!!!")
            game_on = False
        else:
            # INCREMENTO AO TURNO
            turn += 1
    # OPÇÃO PARA CONTINUAR JOGANDO
    game_on = gameon_choice()

clear_output()
```
