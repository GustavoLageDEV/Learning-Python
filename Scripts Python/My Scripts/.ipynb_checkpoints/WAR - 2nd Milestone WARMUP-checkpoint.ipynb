```python
import random
from IPython.display import clear_output


suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card():
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]

    def __str__(self):
        return f"{self.rank} of {self.suit}"
```

```python
class Deck():
    
    def __init__(self):
        
        self.all_cards = []

        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))

    def shuffle_deck(self):
        
        random.shuffle(self.all_cards)

    def deal_one(self):
        return self.all_cards.pop()
```

```python
class Player():

    def __init__(self, name):
        self.name = name
        self.all_cards = []

    def remove_one(self):
        return self.all_cards.pop(0)

    def add_cards(self,new_cards):
        if type(new_cards) == type([]):
            # add multiple card objects
            return self.all_cards.extend(new_cards)
        else:
            # add a single card object
            return self.all_cards.append(new_cards)

    def __str__(self):
        return f"Player {self.name} has {len(self.all_cards)} cards."
```

```python
# WAR CARD GAME LOGIC
# GAME SETUP
player_one = Player("One")
player_two = Player("Two")

new_deck = Deck()
new_deck.shuffle_deck()
game_on = True

while len(new_deck.all_cards) != 0:
    player_one.add_cards(new_deck.deal_one())
    player_two.add_cards(new_deck.deal_one())

round_num = 0

while game_on:

    # ROUND COUNTER
    round_num += 1
    print(f"\nRound {round_num}\n")

    # PRINTING NUM OF CARDS OF EACH PLAYER
    print(player_one)
    print(player_two)
    
    # CHECK WINNER
    if len(player_one.all_cards) == 0:
        print("PLayer 1 is out of cards! Player 2 WINS!")
        game_on = False 
        break
    elif len(player_two.all_cards) == 0:
        print("PLayer 2 is out of cards! Player 1 WINS!")
        game_on = False
        break

    # SHUFFLE BOTH PLAYERS DECKS
    random.shuffle(player_one.all_cards)
    random.shuffle(player_two.all_cards)
    
    # NEW ROUND STARTS
    player_one_table = []
    player_one_table.append(player_one.remove_one())
    
    player_two_table = []
    player_two_table.append(player_two.remove_one())

    # WHILE AT WAR
    war = True
    while war:
        
        # CHECKING FOR GREATER CARD
        if player_one_table[-1].value > player_two_table[-1].value:
            print(f"Player 1 card: {player_one_table[-1]} is greater than PLayer 2 card: {player_two_table[-1]}")
            print(f"Player 1 won round {round_num}")
            player_one.add_cards(player_one_table)
            player_one.add_cards(player_two_table)
            war = False
        elif player_one_table[-1].value < player_two_table[-1].value:
            print(f"PLayer 2 card:  {player_two_table[-1]} greater than Player 1 card: {player_one_table[-1]}")
            print(f"Player 2 won round {round_num}")
            player_two.add_cards(player_one_table)
            player_two.add_cards(player_two_table)
            war = False

        # ACTUALLY WAR APPENS
        else:
            print(f"\nWARRR!! Cards equals {player_two_table[-1]} = {player_one_table[-1]}")
            
            #CHECKING FOR < 5 CARDS ON A PLAYERS' DECK (LOSE IF TRUE)
            if len(player_two.all_cards) < 5:
                print("PLayer 2 don't have cards for WAR. Player 1 WINS!")
                game_on = False
                break
            elif len(player_one.all_cards) < 5:
                print("PLayer 1 don't have cards for WAR. Player 2 WINS!")
                game_on = False
                break
            else:
                # + 5 CARDS TO WAR PILE
                for mun in range(5):
                    player_one_table.append(player_one.remove_one())
                    player_two_table.append(player_two.remove_one())
                    
            print("+ 5 cards to each pile\n")

print("\nGAME OVER!")
```
